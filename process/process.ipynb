## Process 15 minutes interval data to one hour interval data
take average of the four slots 

```python
import pandas as pd
import csv
```

```python
toWrite = "Germany_july25.csv"
```

```python
toRead = "sample/Germany_july25_15min.csv"
processdata = pd.read_csv(toRead)
processdata = processdata.iloc[:, 2:]
processdata = processdata.replace('n/e', 0)
processdata = processdata.replace('N/A', 0)
processdata = processdata.replace('-', 0)

processdata.shape
```

```python
#Create new columm, with one hour time interval
header = dict()
timeInterval = [] 
countryCode = []
with open(toRead, 'r') as file: 
    file_obj = csv.reader(file)
    firstrow = next(file_obj)
    for index, column in enumerate(firstrow): 
            header[column.strip()] = index


    for index, row in enumerate(file_obj, start=0):
        if(index %4 == 0):
            
            prev = index
            firsthalf = row[header["MTU"]].split(' - ')[0]
        if(index == prev+3): 
            secondhalf = row[header["MTU"]].split(' - ')[1]
            interval = "%s - %s"%(firsthalf, secondhalf)#firsthalf, " - ",secondhalf
            # print(interval)
            timeInterval.append(interval)
            countryCode.append(row[header["Area"]])


processdata.head()
```

```python


processdata_combine = processdata.groupby(processdata.index//4).mean()

processdata_combine
```

```python
processdata_combine.insert(0, "Area",countryCode)
processdata_combine.insert(1, "MTU", timeInterval)
processdata_combine
```

```python
processdata_combine.to_csv(toWrite, encoding='utf-8', index=False)
```
